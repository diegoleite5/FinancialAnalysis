```python
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
fim = 'https://www.fundsexplorer.com.br/ranking'
```

```python
try:
    req = Request(fim, headers={'User-Agent':'Mozzila/5.0'})
    response = urlopen(req)
    html = response.read()
    print(html)
except HTTPError as e:
    print(e.status, e.reason)
except URLError as e:
    print(e.reason)
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
table = soup.find('table', {'id': 'table-ranking'})
```

```python
lines = table.find_all('tr')
```

```python
data = []
for line in lines:
    columns = line.find_all('td')
    
    data_line = []
    for col in columns:
        data_line.append(col.text)
        
    data.append(data_line)
```

```python
table_head = [thead.text for thead in table.find('thead').find_all('th')]
```

```python
df = pd.DataFrame(data=data, columns=table_head)
```

```python
today = datetime.today().strftime('%d-%m-%y')
```

```python
df.to_csv(f'Data/{today}.csv')
```
